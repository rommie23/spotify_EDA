```python
corr_df = df_track.drop(["key","mode","explicit"], axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap= sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=1, vmax=1, center=0, cmap="rocket", linewidths=1, linecolor="Black")
heatmap.set_title("Correlation HeatMap Between Variables")
#heatmap.xticklables(heatmap.get_xticklabels(), rotation=90)
```

```python
# using 0.4% sample of df and making regression plots

sample_df = df_track.sample(int(0.004*len(df_track)))
```

```python
len(sample_df)
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data= sample_df, y= "loudness", x="energy", color = "c").set(title = "Loudness and Energy Correlation")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data= sample_df, y= "popularity", x="acousticness", color = "b").set(title = "Popularity and Energy Acousticness")
```

```python
# create new column of year of releasing date

df_track["dates"]=df_track.index.get_level_values("release_date")
df_track.dates = pd.to_datetime(df_track.dates)
years=df_track.dates.dt.year
```

```python
# sns.__version__ = 0.11.2

sns.displot(data = years, discrete = True, aspect=2, height=6, kind="hist").set(title="Number of songs per year")
```

```python
# find duration of songs 
duration_df = df_track.duration
fig_dims = (18,6)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = duration_df, ax=ax, errwidth=False).set(title= "Year VS Duration")
plt.xticks(rotation=90)
```

```python
df_genre = pd.read_csv(r"C:\Users\hp\Desktop\Data Analysis\spotify_project\spotify\SpotifyFeatures.csv")
df_genre.head()
```

```python
plt.title("Duration of songs by genre")
sns.color_palette(palette = "rocket", as_cmap=True)
sns.barplot(data = df_genre, x= "duration_ms", y= "genre")
plt.xlabel("Duration in Miliseconds")
plt.ylabel("Songs Genre")
```

```python
# change miliseconds to seconds by using lambda function
df_genre["duration"] = df_genre["duration_ms"].apply(lambda x: round(x/1000))
plt.figure(figsize=(15,6))
plt.title("Duration of songs by genre")
sns.color_palette(palette = "rocket", as_cmap=True)
sns.barplot(data = df_genre, x= "duration", y= "genre")
plt.xlabel("Duration in Seconds")
plt.ylabel("Songs Genre")
```

```python
sns.set_style(style="darkgrid")
plt.figure(figsize=(10,5))
popular = df_genre.sort_values("popularity", ascending=False).head(10)
sns.barplot(data = popular, x= "popularity", y="genre")
plt.title("Top 5 Genre by Popularity")
```
